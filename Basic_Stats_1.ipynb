### BoxPlot

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

for i, ax in enumerate(axs.flat):

    sns.boxplot(df[numeric_cols[i]], ax=ax, color='darkorange')

    Q1 = df[numeric_cols[i]].quantile(0.25)
    Q3 = df[numeric_cols[i]].quantile(0.75)
    IQR = Q3 - Q1
    
    print(f"Insights for {numeric_cols[i]}:")
    print(f"  Median: {df[numeric_cols[i]].median():.2f}")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Whisker: {df[numeric_cols[i]].min():.2f}")
    print(f"  Upper Whisker: {df[numeric_cols[i]].max():.2f}")
    print(f"  Outliers: {df[numeric_cols[i]][(df[numeric_cols[i]] < (Q1 - 1.5 * IQR)) | (df[numeric_cols[i]] > (Q3 + 1.5 * IQR))].count()}")
    print()

    ax.annotate(
        f'IQR: {IQR:.2f}', 
        xy=(0.5, 1.05), 
        xycoords='axes fraction',
        fontsize=8,
        bbox=dict(facecolor='yellow')
    )
    
    ax.set_title(numeric_cols[i],loc='left')

plt.tight_layout()

plt.show()
```

 ## Conversion of Categorical Data into Dummy Variables

Converting categorical data into dummy variables transforms each category into binary columns (0 or 1). This is crucial because most machine learning models require numerical input. Dummy variables prevent misinterpretation by treating each category independently, ensuring accurate model training and performance.

```python
df.columns
```

```python
col_for_encoding = [ 'Day', 'SKU','BU', 'Brand', 'Model',] # these columns required for analysis not all are required that's why i encoded only these columns
```

```python
df_one_hot_encoded = pd.get_dummies(df,columns=col_for_encoding)
```

```python
bool_columns = df_one_hot_encoded.select_dtypes(include=['bool'])
for i in bool_columns:
    
    df_one_hot_encoded[i] = df_one_hot_encoded[i].astype(int)
    
df_one_hot_encoded.head()
```
